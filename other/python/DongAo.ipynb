```python
from PIL import Image

cn_image_heigh=2605
cn_image_width=1800

def cn_rgb_count_dict(url,cn_rgb=70):
    print(url)
    im=Image.open('dongao/step1/'+url)
    pix=im.load()
    result={}
    for heigh in range(im.size[1]):
        count=0
        for width in range(im.size[0]):
            temp=pix[width,heigh]
            if(type(temp)==tuple):
#                 print(temp)
                temp=temp[0]
            if(temp<cn_rgb):
                count+=1
            else:
                count+=0
        result[heigh]=count
#     print(result,im.size[0],im.size[1])
    return result,im.size[0],im.size[1]



def cn_find_top(cn_now_row,cn_rgb_count_dict,cn_static_mode):
    while(cn_static_mode=='find_top' or cn_static_mode=='begin'):
        if(cn_rgb_count_dict[cn_now_row]!=0):
            cn_static_top.append(cn_now_row)
            cn_static_mode='find_bottom'
        cn_now_row+=1
        if(cn_now_row>int(cn_image_heigh-1)):
            cn_static_mode='end'
    return cn_now_row,cn_static_mode
 
    
def cn_find_bottom(cn_now_row,cn_rgb_count_dict,cn_static_mode,cn_while_max_heigh=40):
    cn_last_black_row=cn_now_row
    while(cn_static_mode=='find_bottom'):
        cn_temp=cn_now_row-cn_last_black_row
        if(cn_rgb_count_dict[cn_now_row]==0 and  cn_temp>cn_while_max_heigh):
            cn_static_bottom.append(cn_last_black_row)
            cn_static_mode='find_top'
        elif(cn_rgb_count_dict[cn_now_row]!=0):
            cn_last_black_row=cn_now_row
        cn_now_row+=1
        if(cn_now_row>int(cn_image_heigh-1)):
            cn_static_bottom.append(cn_now_row)
            cn_static_mode='find_top'
    return cn_now_row,cn_static_mode




def cn_setp2(cn_image_no,url,cn_static_top,cn_static_bottom,cn_static_mode,cn_now_row):

    cn_rgb_count_dict_1,cn_image_width,cn_image_heigh=cn_rgb_count_dict(url)
#     print(cn_rgb_count_dict_1,cn_image_width,cn_image_heigh)
    while(cn_now_row<cn_image_heigh):
        if(cn_static_mode=='find_top' or cn_static_mode=='begin'):
            temp=cn_find_top(cn_now_row,cn_rgb_count_dict_1,cn_static_mode)
            cn_now_row=temp[0]
            cn_static_mode=temp[1]
        elif(cn_static_mode=='find_bottom'):
            temp=cn_find_bottom(cn_now_row,cn_rgb_count_dict_1,cn_static_mode)
            cn_now_row=temp[0]
            cn_static_mode=temp[1]
        
    im=Image.open('dongao/step1/'+url)
    pix=im.load()
    
    for i in range(min(len(cn_static_bottom),len(cn_static_top))):
#         print(cn_static_bottom,cn_static_top)
        if(int(cn_static_bottom[i])-int(cn_static_top[i])<10):
            continue
        box=(0,cn_static_top[i],cn_image_width,cn_static_bottom[i])
 
        region=im.crop(box)
        region.save('dongao/step2/'+str(cn_image_no)+'.jpg')
        cn_image_no=int(cn_image_no)
        cn_image_no+=1
    return cn_image_no

```

```python

cn_image_width=1800

def cn_rgb_count_dict2(url,cn_rgb=70):
    im=Image.open('dongao/step2/'+url)
    pix=im.load()
    result={}
    for width in range(im.size[0]):
        count=0
        for heigh in range(im.size[1]):
            temp=pix[width,heigh]
            if(type(temp)==tuple):
                temp=temp[0]
            if(temp<cn_rgb):
                count+=1
            else:
                count+=0
        result[width]=count
    return result,im.size[0],im.size[1]



def cn_find_left(cn_now_column,cn_rgb_count_dict2,cn_static_mode):
    while(cn_static_mode=='find_left' or cn_static_mode=='begin'):
#         print(len(cn_rgb_count_dict2),cn_now_column)
        if(cn_rgb_count_dict2[cn_now_column]>5):
            cn_static_left.append(cn_now_column)
            cn_static_mode='find_right'
        cn_now_column+=1
#         print(cn_now_column,cn_image_width)
        if(cn_now_column>int(cn_image_width-1)):
            cn_static_mode='end'
    return cn_now_column,cn_static_mode


def cn_find_right(cn_now_column,cn_rgb_count_dict2,cn_static_mode,cn_while_max_width=40):
    cn_last_black_column=cn_now_column
    while(cn_static_mode=='find_right'):
        cn_temp=cn_now_column-cn_last_black_column
        if(cn_rgb_count_dict2[cn_now_column]<=5 and  cn_temp>cn_while_max_width):
            cn_static_right.append(cn_last_black_column)
            cn_static_mode='find_left'
        elif(cn_rgb_count_dict2[cn_now_column]>5):
            cn_last_black_column=cn_now_column
        cn_now_column+=1
        if(cn_now_column>int(cn_image_width-1)):
            cn_static_right.append(cn_now_column)
            cn_static_mode='find_left'
    return cn_now_column,cn_static_mode


def cn_step3(cn_image_no,url,cn_static_left,cn_static_right,cn_static_mode,cn_now_column):

    cn_rgb_count_dict_2,cn_image_width,cn_image_heigh=cn_rgb_count_dict2(url)

    while(cn_now_column<cn_image_width):
        if(cn_static_mode=='find_left' or cn_static_mode=='begin'):
            temp=cn_find_left(cn_now_column,cn_rgb_count_dict_2,cn_static_mode)
            cn_now_column=temp[0]
            cn_static_mode=temp[1]
        elif(cn_static_mode=='find_right'):
            temp=cn_find_right(cn_now_column,cn_rgb_count_dict_2,cn_static_mode)
            cn_now_column=temp[0]
            cn_static_mode=temp[1]
        
    im=Image.open('dongao/step2/'+url)
    pix=im.load()
    for i in range(len(cn_static_left)):
#         print(cn_static_left,cn_static_right)
#         print(i)
        try:
            if(int(cn_static_right[i])-int(cn_static_left[i])<50):
                continue
            box=(cn_static_left[i],0,cn_static_right[i],cn_image_heigh)

            region=im.crop(box)
            region.save('dongao/step3/'+str(cn_image_no)+'.jpg')
            cn_image_no=int(cn_image_no)
            cn_image_no+=1
        except:
            continue
    return cn_image_no
```

```python

def cn_step4(im,cn_rgb=150):
#     im=Image.open('dongao/step3/'+url)
    heigh=im.size[1]
    pix=im.load()
    result=[]
    if(heigh>2400):
        h=2400
        while(h>2):
            count=0
            for w in range(im.size[0]):
                temp=pix[w,h]
                if(type(temp)==tuple):
                    temp=temp[0]
                if(temp<cn_rgb):
                    count+=1
                else:
                    count+=0
            if(count==0):
                result.append(h)
                h=1200
                if(len(result)==2):
                    return result
            h-=1
    elif(heigh>1200):
        h=1200
        while(h>2):
            count=0
            for w in range(im.size[0]):
                temp=pix[w,h]
                if(type(temp)==tuple):
                    temp=temp[0]
                if(temp<cn_rgb):
                    count+=1
                else:
                    count+=0
            if(count==0):
                result.append(h)
                return result
            h-=1
    elif(heigh<=1200):
        return [1,2,3]
    
```
